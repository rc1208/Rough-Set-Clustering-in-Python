```python
target_url = ("https://archive.ics.uci.edu/ml/machine-learning-databases/zoo/zoo.data")
zooData = pd.read_csv(target_url,header=None, prefix = "V")
print(zooData.head())

```

```python
zooData = zooData.drop(zooData.columns[[0, 17]], axis=1) #removing 1st and last column
zooData.head(10)
```

```python
def equiv_class(col_name1, col_name2):
 values_in_col_name1 =  list(zooData[col_name1].unique()) #extracting each unique value in column1
 #print(values_in_col_name1)
 roughness_list = []
 mean_roughness = 0.0
 for val in values_in_col_name1:
    #print("Value",val)
    target_set = zooData[zooData[col_name1] == val].index.tolist() #calculating target set for each unique value in column1
    print("Target Set=", target_set)
    #print(target_set)
    temp_zooData = zooData[zooData[col_name1] == val] #temporary dataframe for each unique value in column1
    #print(temp_zooData)
    arr = list(temp_zooData.groupby([col_name1,col_name2])) #group equiv classes according to column2
    lower_approx = []
    upper_approx = []
    for name,group in arr:
        selected_list = list(group.index)
        #print("name", name)
        print("selected list =",selected_list)
        if(set(selected_list).issubset(target_set)):
            print("yes subset")
            lower_approx.append(selected_list)
        if(set(selected_list) & set(target_set)):
            upper_approx.append(selected_list)
    lower_approx_count = sum(map(len, lower_approx))
    upper_approx_count = sum(map(len, upper_approx))
    print("Lower approx=",lower_approx_count)
    print("Upper approx=",upper_approx_count) 
    roughness = 1 - (lower_approx_count/upper_approx_count)
    print("Roughness=",roughness)
    roughness_list.append(roughness)
 mean_roughness = sum(roughness_list)/float(len(roughness_list))
 print("Mean roughness",mean_roughness)
   
 #return arr


equiv_class('V5', 'V6')
```
